```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians
import folium

# Function to calculate distance between two coordinates
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of Earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Define the coastline coordinates
coastline_lat = 28.56367
coastline_lon = -80.57163

# Assume launch site coordinates are available in launch_sites_df
# Calculate distance between launch site and coastline
launch_sites_df['distance_to_coastline'] = launch_sites_df.apply(
    lambda row: calculate_distance(row['Lat'], row['Long'], coastline_lat, coastline_lon), axis=1)

# Find the closest launch site to the coastline
closest_launch_site = launch_sites_df.loc[launch_sites_df['distance_to_coastline'].idxmin()]

# Add a Marker for the closest launch site to the coastline
closest_launch_site_marker = folium.Marker(
    location=[closest_launch_site['Lat'], closest_launch_site['Long']],
    icon=folium.DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(closest_launch_site['distance_to_coastline']),
    )
)

# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = folium.plugins.MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

# Add Mouse Position and closest launch site marker to the map
site_map.add_child(mouse_position)
site_map.add_child(closest_launch_site_marker)

# Display the map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
import folium
from math import sin, cos, sqrt, atan2, radians

# Define the coordinates of the closest city (Houston, TX)
closest_city_lat = 29.7604
closest_city_lon = -95.3698

# Define the coordinates of a launch site (for example, first launch site in the dataframe)
launch_site_lat = launch_sites_df['Lat'].iloc[0]
launch_site_lon = launch_sites_df['Long'].iloc[0]

# Function to calculate distance between two coordinates
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of Earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Calculate the distance between the launch site and the closest city
distance_to_city = calculate_distance(launch_site_lat, launch_site_lon, closest_city_lat, closest_city_lon)

# Create a marker for the closest city with the distance information
city_marker = folium.Marker(
    location=[closest_city_lat, closest_city_lon],
    icon=folium.DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_to_city),
    )
)

# Add the city marker to the map
site_map.add_child(city_marker)

# Draw a line between the launch site and the closest city
line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [closest_city_lat, closest_city_lon]],
    weight=1
)

# Add the line to the map
site_map.add_child(line)

# Display the map
site_map
```
